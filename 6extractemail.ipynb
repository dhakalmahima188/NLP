```python
source_nlp=spacy.load("en_core_web_sm")
nlp=spacy.blank('en')
nlp.add_pipe("ner", source=source_nlp)
nlp.pipe_names
doc = nlp("The quick tesla 6 $ brown fox jumps over the lazy dog")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
