```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
def gradient_descent(x,y):
    lowest_cost=1000000000
    m,b=0,0
    m_cur,b_cur=0,0
    n=len(x)
    iterations=10000
    lr=0.001
    for i in range(iterations) :
        y_pred=m_cur*x + b_cur
        p1= - (2/n)*sum(x*(y-(y_pred)))
        p2=- (2/n)*sum((y-(y_pred)))
        cost=(1/n)*sum([val**2 for val in (y-(y_pred))])
        m_cur=m_cur-lr*p1
        b_cur=b_cur-lr*p2
        
        if cost<lowest_cost:
            lowest_cost=cost
            m=m_cur
            b=b_cur
        
        #print(f"p1={p1} p2={p2} m={m_cur},b={b_cur},cost={cost},iteration={i}")
    print(f"m={m},b={b}, cost={lowest_cost}")
        
  
gradient_descent(x,y)
```
