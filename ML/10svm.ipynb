```python
#svm
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
import pandas as pd
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
df['target']=iris.target
df[df.target==1].head()
```

```python
iris.target_names
```

```python
df['target'].value_counts()
```

```python
df
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
```

```python
df
```

```python
df
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
df2
# Assuming 'df' is the original DataFrame containing the iris data
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
df2.head()


```

```python
import matplotlib.pyplot as plt
plt.plot(df0['sepal length (cm)'],df0['sepal width (cm)'],'r.')
plt.plot(df1['sepal length (cm)'],df1['sepal width (cm)'],'g.')
plt.plot(df2['sepal length (cm)'],df2['sepal width (cm)'],'b.')
```

```python
x=df.drop(['target','flower_name'],axis='columns')
y=df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear',C=30,gamma='auto')
model.fit(x_train, y_train)
```

```python


model.score(x_test,y_test)
```

```python
#exercise
from sklearn.datasets import load_digits
digits=load_digits()
dir(digits)
```

```python
digits.data[0]
```

```python
digits.feature_names
```

```python
df=pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
df['target']=digits.target
```

```python
df
```

```python
x=df.drop('target',axis='columns')
y=df.target 
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
model=SVC(kernel='linear',C=30,gamma='auto')

model.fit(x_train, y_train)
model.score(x_test,y_test)
```
