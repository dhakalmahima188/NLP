```python
data={
  "Area": [
    548.813504, 715.189366, 602.763376, 544.883183, 423.654799
  ],
  "Price": [
    27149.387736, 35984.674940, 30254.584414, 26860.098228, 21554.803015
  ]
}
```

```python
import pandas as pd
df=pd.DataFrame(data)
# x=df[['Area']]
X=df['Area'].values.reshape(-1,1)
X
```

```python
#linear model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,df['Price'])
```

```python
model.predict([[3300]])
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.Area,df.Price,color='red',marker='+')
```

```python
plt.plot(df.Area,model.predict(df[['Area']]),color='blue')
```

```python
data2={
  "Area": [
    548.813504, 715.189366, 602.763376, 544.883183, 423.654799
  ],
  "Price": [
  ]
}
```

```python
for i in range(0,5):
    data2['Price'].append(model.predict(([[data2['Area'][i]]]))[0])
```

```python
data2

df2=pd.DataFrame(data2)
```

```python

plt.plot(df2.Area,df2.Price,color='green',marker='*')
plt.plot(df.Area,df.Price,color='red',marker='+')    
plt.show()
```
